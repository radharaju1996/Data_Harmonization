```python
# Perform detailed missing value analysis and outlier detection.

# Detailed Missing Value Analysis
def analyze_missing_values(df):
    missing_info = pd.DataFrame({
        'Column Name': df.columns,
        'Missing Count': df.isnull().sum(),
        'Missing Percentage': (df.isnull().sum() / len(df)) * 100
    })
    return missing_info

missing_values_df = analyze_missing_values(final)
print("Missing Value Analysis:")
print(missing_values_df)


# Outlier Detection using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers


# replace with actual numerical column names
numerical_columns = ['Temprature_y', 'Humidity_x', 'Celsius_y', 'Count_x']

for col in numerical_columns:
  outliers = detect_outliers_iqr(final, col)
  print(f"\nOutliers in '{col}':")
outliers

```

```python
def analyze_skewness(df):
    # numerical columns
    numerical_df = df.select_dtypes(include=['number'])
    skewness_df = pd.DataFrame({
        'Column Name': numerical_df.columns,
        'Skewness': numerical_df.skew()
    })
    return skewness_df

skewness_info = analyze_skewness(final)
print("Skewness Analysis:")
print(skewness_info)

# Apply transformations based on skewness
for col in final.select_dtypes(include=['number']).columns:
    if abs(final[col].skew()) > 0.5:  # Threshold for skewness (adjust as needed)
      print(f"Applying log transformation to '{col}'")
      final[col] = np.log1p(final[col])

# Print updated skewness after transformation
skewness_info_after = analyze_skewness(final)
print("\nSkewness Analysis After Transformation:")
skewness_info_after

#Insights Gained:


#Identification of Skewed Columns: By calculating the skewness of each numerical column, the code helps identify columns with a significant skew (greater than 0.5 in this case).
#Skewness Reduction: Applying log transformation to the skewed columns aims to reduce their skewness and make their distribution more symmetrical.
#This can be beneficial for some machine learning algorithms that assume normally distributed data.
#Impact of Transformation: By comparing the skewness values before and after the transformation,
# The effectiveness of the transformation in reducing skewness. The skewness values after transformation should be closer to 0, indicating a more symmetrical distribution.
```

```python
# Numerical columns for the histplot
numerical_cols = final.select_dtypes(include=['number']).columns

# Number of rows and columns for subplots
num_cols = 3  # You can adjust the number of columns
num_rows = (len(numerical_cols) + num_cols - 1) // num_cols

plt.figure(figsize=(15, 5 * num_rows))  # Adjust figure size as needed

for i, col in enumerate(numerical_cols):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.histplot(final[col], kde=True)  # Use kde for density estimation
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()


# Insights Gained:

# Shape: The shape of the histogram tells about the data's distribution.
# Symmetrical (bell-shaped): Indicates a normal distribution.
# Skewed (longer tail on one side): Shows that the data is concentrated on one side and has a tail on the other.
# Uniform : Suggests that the data is evenly spread across the range.
# Central Tendency: The peak of the histogram gives you an idea of the most common values (mode) for the feature.
# Spread: The width of the histogram shows how spread out the data is. A wider histogram indicates more variability.
# Outliers: Unusual or extreme values might be visible as isolated bars far from the main body of the histogram.
# KDE: The KDE line provides a smoother estimate of the probability density function, helping to visualize the overall shape of the distribution.
```
